## 銘柄選定
 - 出来高10億円以上の銘柄のみ

```python
%%sql
select
 *
from backtest_result
where 
entry_strategy like 'BreakoutKC[15,0.5][15,0.5]'
and exit_strategy like 'Contract%'

order by rate_of_return desc
```

```python
%%sql
select
 symbol
,substr(text(time), 0, 8)
,sum(profit_rate)
from backtest_history
where symbol in (
select
 symbol
from backtest_history
where date(time) between '2019-01-01' and '2019-12-31'
group by symbol, substr(text(time), 0, 8)
having sum(profit_rate) > 0.1 and substr(text(time), 0, 8) = '2019-12'
)
group by symbol,substr(text(time), 0, 8)
order by 1, 2
```
